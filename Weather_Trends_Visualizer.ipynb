```python
import pandas as pd

df = pd.read_csv("sample_weather.csv", parse_dates=['Date'])
print(df)
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Date'], df['Temperature'])
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperature Trend')
plt.grid(True)
plt.show()
```

```python
plt.bar(df["Date"], df["Temperature"])
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature (Bar Chart)")
plt.grid(axis="y")
plt.show()
```
